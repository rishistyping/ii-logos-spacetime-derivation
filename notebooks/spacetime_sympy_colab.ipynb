# A Brief Derivation of Spacetime — SymPy Companion

This notebook mirrors the v0 matrix spine. It is a computed/explainer surface, not proof authority.

```python
import sympy as sp
lambda_, x = sp.symbols('lambda x', real=True)
A = sp.Matrix([[0, -lambda_], [-1, 0]])
A
```

```python
sp.trace(A), A.det(), sp.expand(A.charpoly(x).as_expr()), sp.simplify(A*A - lambda_*sp.eye(2))
```

Branch examples: `lambda=1`, `lambda=0`, `lambda=-1`.

```python
for val in [1, 0, -1]:
    Av = A.subs(lambda_, val)
    print('lambda =', val, 'det =', Av.det(), 'charpoly =', sp.expand(Av.charpoly(x).as_expr()), 'eigenvalues =', list(Av.eigenvals().keys()))
```
